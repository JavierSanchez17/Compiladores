```python
import re
```

```python
def tokenize(text):
    tokens = []
    # Expresiones regulares para cada tipo de token
    keywords = re.compile(r'\b(if|else|while)\b')
    identificador = re.compile(r'[a-zA-z][a-zA-Z0-9]*')
    numeros = re.compile(r'\d+')
    operadores = re.compile(r'[\+\-\*\/]')
    delimitador = re.compile(r'[()]|;')
    
    # Extraer palabra por palabra
    for word in text.split():
        if keywords.match(word):
            tokens.append(('KeyWord', word))
        elif identificador.match(word):
            tokens.append(('Identificador', word))
        elif numeros.match(word):
            tokens.append(('Numero', word))
        elif operadores.match(word):
            tokens.append(('Operador', word))
        elif delimitador.match(word):
            tokens.append(('Delimitador', word))
        else:
            tokens.append(('Token Desconocido', word))
    return tokens
```

```python
# Ejemplo de uso
texto = 'if x > 5 then ( y = t + 1 ) ; else z = z - 1 ;'
tokens = tokenize(texto)
for i in tokens:
    print(i)
```
